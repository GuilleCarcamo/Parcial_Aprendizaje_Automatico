# Celda 1 — Importación y configuración

```python
# CELDA 1 — IMPORTACIÓN Y CONFIGURACIÓN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    roc_curve, confusion_matrix, classification_report
)
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Estilo de gráficos
plt.rcParams["figure.figsize"] = (7, 5)
plt.rcParams["figure.dpi"] = 120

# Reproducibilidad
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# Ruta del dataset (este notebook vive en notebooks/)
DATA_PATH = Path("../data/raw/structured_endometriosis_data.csv")
print("Ruta del dataset:", DATA_PATH)
```

# Celda 2 — Carga del dataset

```python
# CELDA 2 — CARGA DEL DATASET
df = pd.read_csv(DATA_PATH)
print("Dataset cargado correctamente.")
print("Forma:", df.shape)
display(df.head())

TARGET = "Diagnosis"   # 0 = No, 1 = Sí
assert TARGET in df.columns, "No se encontró la columna objetivo."

print("\nDistribución de Diagnosis (%):")
display((df[TARGET].value_counts(normalize=True) * 100).round(2))
```

# Celda 3 — Preparación (imputación + codificación)

```python
# CELDA 3 — PREPARACIÓN
data = df.copy()

# Separar numéricas y categóricas
num_cols = data.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols = data.select_dtypes(exclude=["int64", "float64"]).columns.tolist()
if TARGET in num_cols: num_cols.remove(TARGET)
if TARGET in cat_cols: cat_cols.remove(TARGET)

# Imputación simple
for c in num_cols:
    data[c] = data[c].fillna(data[c].median())
for c in cat_cols:
    moda = data[c].mode(dropna=True)
    data[c] = data[c].fillna(moda.iloc[0] if len(moda) else "Desconocido")

# One-hot encoding
X = pd.get_dummies(data.drop(columns=[TARGET]), drop_first=True)
y = data[TARGET].astype(int)

print("X:", X.shape, "| y:", y.shape)
```

# Celda 4 — División en entrenamiento y prueba

```python
# CELDA 4 — DIVISIÓN TRAIN/TEST
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, stratify=y, random_state=RANDOM_STATE
)

print("Tamaño train:", X_train.shape, "| tamaño test:", X_test.shape)
print("Estrato train:", y_train.value_counts(normalize=True).round(3).to_dict())
print("Estrato test :", y_test.value_counts(normalize=True).round(3).to_dict())
```

# Celda 5 — Funciones auxiliares (métricas y gráficos)

```python
# CELDA 5 — FUNCIONES AUXILIARES

def mostrar_metricas(y_true, y_pred, y_score):
    """Calcula e imprime métricas básicas."""
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    roc  = roc_auc_score(y_true, y_score)
    print(f"Accuracy:  {acc:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"Recall:    {rec:.3f}")
    print(f"F1-score:  {f1:.3f}")
    print(f"ROC-AUC:   {roc:.3f}")

def mostrar_confusion(y_true, y_pred, titulo):
    """Grafica la matriz de confusión."""
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=["No Endo (0)", "Sí Endo (1)"],
                yticklabels=["No Endo (0)", "Sí Endo (1)"])
    plt.title(titulo)
    plt.xlabel("Predicción")
    plt.ylabel("Real")
    plt.show()

def mostrar_roc(y_true, y_score, titulo):
    """Grafica la curva ROC."""
    fpr, tpr, _ = roc_curve(y_true, y_score)
    plt.plot(fpr, tpr, label="ROC")
    plt.plot([0,1],[0,1],"--", color="gray")
    plt.xlabel("Tasa de falsos positivos")
    plt.ylabel("Tasa de verdaderos positivos")
    plt.title(titulo)
    plt.legend()
    plt.show()
```

# Celda 6 — Random Forest (modelo principal) + métricas + gráficos

```python
# CELDA 6 — RANDOM FOREST
print("=== RANDOM FOREST ===")

rf = RandomForestClassifier(
    n_estimators=300,
    random_state=RANDOM_STATE,
    class_weight="balanced_subsample",
    n_jobs=-1
)
rf.fit(X_train, y_train)

# Predicciones y puntajes
y_pred_rf  = rf.predict(X_test)
y_score_rf = rf.predict_proba(X_test)[:, 1]

# Métricas
mostrar_metricas(y_test, y_pred_rf, y_score_rf)
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred_rf, target_names=["No (0)", "Sí (1)"]))

# Gráficos requeridos
mostrar_confusion(y_test, y_pred_rf, "Random Forest — Matriz de Confusión")
mostrar_roc(y_test, y_score_rf, "Random Forest — Curva ROC")
```

# Celda 7 — Importancia de variables (visual)

```python
# CELDA 7 — IMPORTANCIA DE VARIABLES (Random Forest)
imp = pd.DataFrame({
    "feature": X_train.columns,
    "importance": rf.feature_importances_
}).sort_values("importance", ascending=False)

display(imp.head(20))

# Gráfico de las 15 más importantes
top = imp.head(15).sort_values("importance")
plt.barh(top["feature"], top["importance"])
plt.title("Random Forest — Importancias (Top 15)")
plt.xlabel("Importancia")
plt.ylabel("Variable")
plt.tight_layout()
plt.show()
```

# Celda 8 — Baselines para comparativa (Logística y Árbol)

```python
# CELDA 8 — BASELINES PARA COMPARATIVA
print("=== BASELINES PARA COMPARATIVA (métricas) ===")

# 1) Regresión Logística (con escalado)
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)

log_reg = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
log_reg.fit(X_train_sc, y_train)
y_pred_lr  = log_reg.predict(X_test_sc)
y_score_lr = log_reg.predict_proba(X_test_sc)[:,1]
print("\nRegresión Logística — métricas:")
mostrar_metricas(y_test, y_pred_lr, y_score_lr)

# 2) Árbol de Decisión (sin escalado)
tree = DecisionTreeClassifier(random_state=RANDOM_STATE)
tree.fit(X_train, y_train)
y_pred_dt  = tree.predict(X_test)
y_score_dt = tree.predict_proba(X_test)[:,1]
print("\nÁrbol de Decisión — métricas:")
mostrar_metricas(y_test, y_pred_dt, y_score_dt)
```

# Celda 9 — Comparativa final (tabla en pantalla)

```python
# CELDA 9 — COMPARATIVA FINAL
comparativa = pd.DataFrame([
    {
        "Modelo": "Random Forest",
        "Accuracy":  accuracy_score(y_test, y_pred_rf),
        "Precision": precision_score(y_test, y_pred_rf, zero_division=0),
        "Recall":    recall_score(y_test, y_pred_rf, zero_division=0),
        "F1-score":  f1_score(y_test, y_pred_rf, zero_division=0),
        "ROC-AUC":   roc_auc_score(y_test, y_score_rf)
    },
    {
        "Modelo": "Regresión Logística",
        "Accuracy":  accuracy_score(y_test, y_pred_lr),
        "Precision": precision_score(y_test, y_pred_lr, zero_division=0),
        "Recall":    recall_score(y_test, y_pred_lr, zero_division=0),
        "F1-score":  f1_score(y_test, y_pred_lr, zero_division=0),
        "ROC-AUC":   roc_auc_score(y_test, y_score_lr)
    },
    {
        "Modelo": "Árbol de Decisión",
        "Accuracy":  accuracy_score(y_test, y_pred_dt),
        "Precision": precision_score(y_test, y_pred_dt, zero_division=0),
        "Recall":    recall_score(y_test, y_pred_dt, zero_division=0),
        "F1-score":  f1_score(y_test, y_pred_dt, zero_division=0),
        "ROC-AUC":   roc_auc_score(y_test, y_score_dt)
    }
])

display(comparativa.round(3).sort_values("ROC-AUC", ascending=False))
```
